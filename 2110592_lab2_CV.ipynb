```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the grayscale image
image_path = "/kaggle/input/images-for-cv/img_001.png"
image = cv2.imread(image_path)

if len(image.shape) != 2:
    # Convert the image to grayscale
    image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
# Gamma value for correction
gamma_value = 2.2  # Example gamma value (you can change this)

# Apply Gamma Correction
# Step 1: Normalize pixel values to range [0, 1]
normalized_image = image / 255.0

# Step 2: Apply gamma correction (g(x) = (f(x))^gamma)
gamma_corrected_image = 255 * (normalized_image ** gamma_value)

# Step 3: Clip the values to ensure they're within [0, 255]
gamma_corrected_image = np.clip(gamma_corrected_image, 0, 255).astype(np.uint8)

# Apply Inverse Gamma Correction
# Step 1: Apply inverse gamma correction (g(x) = (f(x))^(1/gamma))
inverse_gamma_corrected_image = 255 * (normalized_image ** (1 / gamma_value))

# Step 2: Clip the values to ensure they're within [0, 255]
inverse_gamma_corrected_image = np.clip(inverse_gamma_corrected_image, 0, 255).astype(np.uint8)

# Plot the original, gamma-corrected, and inverse gamma-corrected images
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title("Gamma Corrected ( = {gamma_value})")
plt.imshow(gamma_corrected_image, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title("Inverse Gamma Corrected ( = 1/{gamma_value})")
plt.imshow(inverse_gamma_corrected_image, cmap='gray')
plt.axis('off')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the grayscale image
image_path = "/kaggle/input/images-for-cv/img_001.png"
image = cv2.imread(image_path)

if len(image.shape) != 2:
    # Convert the image to grayscale
    image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
# Define a list of gamma values to apply
gamma_values = [0.5, 1.0, 1.5, 2.0, 2.5]

# Create a plot to show results
plt.figure(figsize=(15, len(gamma_values) * 5))

# Loop over the gamma values
for i, gamma_value in enumerate(gamma_values):
    
    # Apply Gamma Correction
    normalized_image = image / 255.0
    gamma_corrected_image = 255 * (normalized_image ** gamma_value)
    gamma_corrected_image = np.clip(gamma_corrected_image, 0, 255).astype(np.uint8)

    # Apply Inverse Gamma Correction (gamma = 1 / gamma_value)
    inverse_gamma_corrected_image = 255 * (normalized_image ** (1 / gamma_value))
    inverse_gamma_corrected_image = np.clip(inverse_gamma_corrected_image, 0, 255).astype(np.uint8)

    # Plot the results for each gamma value
    plt.subplot(len(gamma_values), 3, i * 3 + 1)
    plt.title("Original Image")
    plt.imshow(image, cmap='gray')
    plt.axis('off')

    plt.subplot(len(gamma_values), 3, i * 3 + 2)
    plt.title("Gamma Corrected ( = {gamma_value})")
    plt.imshow(gamma_corrected_image, cmap='gray')
    plt.axis('off')

    plt.subplot(len(gamma_values), 3, i * 3 + 3)
    plt.title("Inverse Gamma ( = 1/{gamma_value})")
    plt.imshow(inverse_gamma_corrected_image, cmap='gray')
    plt.axis('off')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

foreground_image_path = "/kaggle/input/images-for-cv/img_001.png"  # Path to your foreground image with alpha
background_image_path = "/kaggle/input/images-for-cv/img_007.png"  

# Load the foreground with an attempt to include the alpha channel
foreground = cv2.imread(foreground_image_path, cv2.IMREAD_UNCHANGED)

# Check if alpha channel is present
if foreground.shape[2] == 4:
    # Split into BGR and alpha channels
    bgr_foreground = foreground[:, :, :3]
    alpha_mask = foreground[:, :, 3]
else:
    # If no alpha, create a fully opaque mask
    bgr_foreground = foreground
    alpha_mask = np.ones(bgr_foreground.shape[:2], dtype=np.uint8) * 255  # Fully opaque

# Load and resize the background
background = cv2.imread(background_image_path)
background = cv2.resize(background, (bgr_foreground.shape[1], bgr_foreground.shape[0]))

# Normalize alpha mask to [0, 1] for blending
alpha = alpha_mask.astype(float) / 255.0

# Convert images to float for blending
bgr_foreground = bgr_foreground.astype(float)
background = background.astype(float)

# Composite the images using alpha blending: C = (1 - alpha) * B + alpha * F
composite_image = (1 - alpha[:, :, None]) * background + alpha[:, :, None] * bgr_foreground
composite_image = composite_image.astype(np.uint8)

# Display results
plt.figure(figsize=(12, 5))

# Original source image
plt.subplot(1, 5, 1)
plt.title("(a) Source Image")
plt.imshow(cv2.cvtColor(bgr_foreground.astype(np.uint8), cv2.COLOR_BGR2RGB))
plt.axis('off')

# Extracted foreground object
plt.subplot(1, 5, 2)
plt.title("(b) Extracted Foreground Object F")
plt.imshow(cv2.cvtColor((alpha[:, :, None] * bgr_foreground).astype(np.uint8), cv2.COLOR_BGR2RGB))
plt.axis('off')

# Alpha matte in grayscale
plt.subplot(1, 5, 3)
plt.title("(c) Alpha Matte (Grayscale)")
plt.imshow(alpha_mask, cmap='gray')
plt.axis('off')

# Background image
plt.subplot(1, 5, 4)
plt.title("Background Image")
plt.imshow(cv2.cvtColor(background.astype(np.uint8), cv2.COLOR_BGR2RGB))
plt.axis('off')

# New composite image
plt.subplot(1, 5, 5)
plt.title("(d) New Composite C")
plt.imshow(cv2.cvtColor(composite_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

# Display all images
plt.tight_layout()
plt.show()
```
